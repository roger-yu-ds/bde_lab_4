```python
def list_bucket_contents(bucket, match=''):
    s3_resource = boto3.resource('s3')
    bucket_resource = s3_resource.Bucket(bucket)
    for key in bucket_resource.objects.all():
        if match in key.key:
            print(key.key)
```

```python
dfs = spark.read.format('csv').options(header='true').load(f"s3a://{bucket_name}/covid19tos3/jh_csse_daily/0_1588*.csv")
```

```python
dfs.count()
```

```python
cols_list = ['date', 
             'country_region_cleansed', 
             'province_state_cleansed', 
             'admin_area', 
             'latitude', 
             'longitude', 
             'confirmed', 
             'active', 
             'deaths', 
             'recovered']
```

```python
dfs_cleaned = dfs.select(cols_list)
```

```python
dfs_cleaned.show(5)
```

# Spark Dataframe

```python
dfs_cleaned = dfs_cleaned.withColumn('confirmed', F.col('confirmed').astype(IntegerType())).\
    withColumn('active', F.col('active').astype(IntegerType())).\
    withColumn('deaths', F.col('deaths').astype(IntegerType())).\
    withColumn('recovered', F.col('recovered').astype(IntegerType()))
```

```python
dfs_cleaned.printSchema()
```

```python
dfs_cleaned.select('date').first()
```

```python
dfs_cleaned.filter(F.col('province_state_cleansed').rlike('Carolina|Washington')).show()
```

```python
dfs_cleaned.groupBy('country_region_cleansed').max('deaths').show()
```

```python
dfs_country = dfs_cleaned.groupBy('country_region_cleansed').\
    agg(F.sum('active').alias('total_active')).\
    filter(F.col('total_active') >= 100000).\
    sort(F.col('total_active').desc())

dfs_country.show()
```

```python
dfs_pivot = dfs_cleaned.filter(F.col('country_region_cleansed') == 'United States of America').\
    groupBy('province_state_cleansed').agg(
    F.sum('active').alias('total_active')).\
    filter(F.col('total_active') >= 100000).\
    sort(F.col('total_active').desc()
)
```
